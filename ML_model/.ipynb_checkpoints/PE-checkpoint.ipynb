### 4. Раскрытие признаков

```python
def iv_woe(data, target, bins=10, show_woe=False):
    # Пустые датафреймы
    newDF,woeDF = pd.DataFrame(), pd.DataFrame()
    
    # Названия всех колонок
    cols = data.columns
    
    # Считаем WOE и IV для всех независимых переменных
    for ivars in cols[~cols.isin([target])]:
        if (data[ivars].dtype.kind in 'bifc') and (len(np.unique(data[ivars]))>10):
            binned_x = pd.qcut(data[ivars], bins,  duplicates='drop')
            d0 = pd.DataFrame({'x': binned_x, 'y': data[target]})
        else:
            d0 = pd.DataFrame({'x': data[ivars], 'y': data[target]})
        d = d0.groupby("x", as_index=False).agg({"y": ["count", "sum"]})
        d.columns = ['Cutoff', 'N', 'Events']
        d['% of Events'] = np.maximum(d['Events'], 0.5) / d['Events'].sum()
        d['Non-Events'] = d['N'] - d['Events']
        d['% of Non-Events'] = np.maximum(d['Non-Events'], 0.5) / d['Non-Events'].sum()
        d['WoE'] = np.log(d['% of Events']/d['% of Non-Events'])
        d['IV'] = d['WoE'] * (d['% of Events'] - d['% of Non-Events'])
        d.insert(loc=0, column='Variable', value=ivars)
        print("Information value of " + ivars + " is " + str(round(d['IV'].sum(),6)))
        temp =pd.DataFrame({"Variable" : [ivars], "IV" : [d['IV'].sum()]}, columns = ["Variable", "IV"])
        newDF=pd.concat([newDF,temp], axis=0)
        woeDF=pd.concat([woeDF,d], axis=0)

        # Показать таблицу WoE
        if show_woe == True:
            print(d)
    return newDF, woeDF
```

```python
dataset.legitimate = dataset.legitimate.astype('int64')
```

### 5. Data Imbalance - SMOTE TOMEK

```python
X=dataset[features]
y=dataset['legitimate']
```

```python
randomseed = 42
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
print(X_test.shape[0] + X_train.shape[0])
print('Объем тренировочной выборки (целевая переменная):', y_train.shape)
print('Объем тестовой выборки (целевая переменная):', y_test.shape)
print('Объем тренировочной выборки (признаки)::', X_train.shape)
print('Объем тестовой выборки (целевая переменная):', X_test.shape)
```

```python
"""
Перед SMOTE_Tomek
"""
counter_train = Counter(y_train)
counter_test = Counter(y_test)
print(counter_train, counter_test)



over_and_under_sample =  imblearn.combine.SMOTETomek(sampling_strategy = 1.0, n_jobs = -1, random_state = randomseed)
X_train, y_train = over_and_under_sample.fit_resample(X_train, y_train)

counter_train = Counter(y_train)
counter_test = Counter(y_test)
print(counter_train, counter_test)
```

```python
print(X_test.shape[0] + X_train.shape[0])
print('Training labels shape:', y_train.shape)
print('Test labels shape:', y_test.shape)
print('Training features shape:', X_train.shape)
print('Test features shape:', X_test.shape)
```

### 6. SVM

```python
clf = LinearSVC()
clf.fit(X_train,y_train)    
y_pred = clf.predict(X_test)
```

### 7. Результаты обучения модели методом опорных векторов

#### Матрица ошибок

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
classes = ['Safe', 'Risky/Ransomware']

figure, ax = plot_confusion_matrix(conf_mat = cm,
                                   class_names = classes,
                                   colorbar = True)

plt.show()
```

#### Основные метрики

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
print(f" - Accuracy (доля правильных ответов): {accuracy}")
print(f" - Precision (точность): {precision}")
print(f" - Recall (полнота): {recall}")
print(f" - F1-score: {f1}")
```

#### AUC-ROC

```python
auc = np.round(roc_auc_score(y_test, y_pred), 3) 
print(f"AUC-ROC: {auc}")
```

### 8. LazyPredict (попробуем найти лучшую модель)

```python
from sklearn.utils import all_estimators
from sklearn.base import ClassifierMixin
import xgboost
import lightgbm

removed_classifiers = [
    "ClassifierChain",
    "ComplementNB",
    "GradientBoostingClassifier",
    "GaussianProcessClassifier",
    "HistGradientBoostingClassifier",
    "MLPClassifier",
    "LogisticRegressionCV", 
    "MultiOutputClassifier", 
    "MultinomialNB", 
    "OneVsOneClassifier",
    "OneVsRestClassifier",
    "OutputCodeClassifier",
    "RadiusNeighborsClassifier",
    "VotingClassifier",
    "NuSVC" # Удален, т.к. вызывает вечный цикл
]

CLASSIFIERS = [
    est
    for est in all_estimators()
    if (issubclass(est[1], ClassifierMixin) and (est[0] not in removed_classifiers))
]

CLASSIFIERS.append(("XGBClassifier", xgboost.XGBClassifier))
CLASSIFIERS.append(("LGBMClassifier", lightgbm.LGBMClassifier))
```

```python
cl = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None, classifiers = CLASSIFIERS)
models,predictions = cl.fit(X_train, X_test, y_train, y_test)

print(models)
```
