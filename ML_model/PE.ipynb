```python
import os
import pandas
import numpy as np
import pickle
import pefile
import sklearn.ensemble as ek
from sklearn.feature_selection import SelectFromModel
import joblib
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn import svm
import sklearn.metrics as metrics
```

#### Выборка признаков

```python
# Признаки
X = dataset.drop(['Name','md5','legitimate'],axis=1).values    # Отбрасываем лишние данные (используем для обучения только float и int)
# Целевая переменная
y = dataset['legitimate'].values
```

#### Подгонка данных и выбор важных переменных

```python
extratrees = ek.ExtraTreesClassifier().fit(X,y)
model = SelectFromModel(extratrees, prefit=True)
X_new = model.transform(X)
nbfeatures = X_new.shape[1]
```

```python
# Количество важных признаков
nbfeatures
```

```python
# разделяем данные (70% - тренировочная и 30% - тестовая)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_new, y ,test_size=0.3, stratify = y)
```

```python
features = []
index = np.argsort(extratrees.feature_importances_)[::-1][:nbfeatures]
```

#### Все необходимые признаки:

```python
for f in range(nbfeatures):
    print("%d. feature %s (%f)" % (f + 1, dataset.columns[2+index[f]], extratrees.feature_importances_[index[f]]))
    features.append(dataset.columns[2+f])
```

#### Тестируем, какой классификатор покажет лучшие результаты:

```python
model = { "Decision Tree": DecisionTreeClassifier(max_depth=10),
         "Random Forest":ek.RandomForestClassifier(n_estimators=50)}
```

```python
results = {}
score = 0

for algo in model:
    
    clf = model[algo]
    clf.fit(X_train,y_train)
    
    if clf.score(X_test, y_test) > score:
        y_pred = clf.predict(X_test)
    score = clf.score(X_test,y_test)
    print ("%s : %s " %(algo, score))
    results[algo] = score
```

```python
winner = max(results, key=results.get)# Выбираем классификатор с наилучшим результатом
print("Using", winner, "for classification, with",len(features), 'main features.')

    
        
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])    
plt.figure(figsize=(4, 3))
sns.heatmap(confusion_matrix, annot=True, cmap='Blues', fmt='d')
plt.title(f'Confusion Matrix for {algo}')
plt.show()
```

#### Сохраняем модель как classifier.pkl, и признаки как features.pkl

```python
joblib.dump(model[winner],'..\\Classifier\\classifier.pkl')
open('..\\Classifier\\features.pkl', 'wb').write(pickle.dumps(features))
```
